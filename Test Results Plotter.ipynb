```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()


```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
sgd = pd.read_csv('/content/drive/My Drive/DL_PLOT_CSV/OneLayerPerceptron/1_layer_Perceptron(SGD).csv')
adagrad = pd.read_csv('/content/drive/My Drive/DL_PLOT_CSV/OneLayerPerceptron/1_layer_Perceptron(AdaGrad).csv')
adam = pd.read_csv('/content/drive/My Drive/DL_PLOT_CSV/OneLayerPerceptron/1_layer_Perceptron(Adam).csv')
amsgrad = pd.read_csv('/content/drive/My Drive/DL_PLOT_CSV/OneLayerPerceptron/1_layer_Perceptron(AMSGrad).csv')
adabound = pd.read_csv('/content/drive/My Drive/DL_PLOT_CSV/OneLayerPerceptron/1_layer_Perceptron(AdaBound).csv')
amsbound = pd.read_csv('/content/drive/My Drive/DL_PLOT_CSV/OneLayerPerceptron/1_layer_Perceptron(AmsBound).csv')
```

```
fig,axs = plt.subplots(2,2,figsize=(13,9))

fig.suptitle('Results for One Layer Perceptron',fontsize=16)
axs[0,0].plot(range(100),sgd['train_acc'],label='sgd')
axs[0,0].plot(range(100),adagrad['train_acc'],label='adagrad')
axs[0,0].plot(range(100),adam['train_acc'],label='adam')
axs[0,0].plot(range(100),amsgrad['train_acc'],label='amsgrad')
axs[0,0].plot(range(100),adabound['train_acc'],label='adabound')
axs[0,0].plot(range(100),amsbound['train_acc'],label='amsbound')

axs[0,0].set_title('train accuracy')
axs[0,0].legend(loc='lower right')

#############################################


axs[0,1].plot(range(100),sgd['val_acc'],label='sgd')
axs[0,1].plot(range(100),adagrad['val_acc'],label='adagrad')
axs[0,1].plot(range(100),adam['val_acc'],label='adam')
axs[0,1].plot(range(100),amsgrad['val_acc'],label='amsgrad')
axs[0,1].plot(range(100),adabound['val_acc'],label='adabound')
axs[0,1].plot(range(100),amsbound['val_acc'],label='amsbound')
axs[0,1].set_title('test accuracy')
axs[0,1].legend(loc='lower right')


#########################################################
axs[1,0].plot(range(100),sgd['train_loss'],label='sgd')
axs[1,0].plot(range(100),adagrad['train_loss'],label='adagrad')
axs[1,0].plot(range(100),adam['train_loss'],label='adam')
axs[1,0].plot(range(100),amsgrad['train_loss'],label='amsgrad')
axs[1,0].plot(range(100),adabound['train_loss'],label='adabound')
axs[1,0].plot(range(100),amsbound['train_loss'],label='amsbound')


axs[1,0].set_title('train loss')
axs[1,0].legend(loc='upper right')



##############################################

axs[1,1].plot(range(100),sgd['val_loss'],label='sgd')
axs[1,1].plot(range(100),adagrad['val_loss'],label='adagrad')
axs[1,1].plot(range(100),adam['val_loss'],label='adam')
axs[1,1].plot(range(100),amsgrad['val_loss'],label='amsgrad')
axs[1,1].plot(range(100),adabound['val_loss'],label='adabound')
axs[1,1].plot(range(100),amsbound['val_loss'],label='amsbound')


axs[1,1].set_title('val loss')
axs[1,1].legend(loc='upper right')
plt.savefig('1_layer_Perceptron_result_plot.png')
plt.show()
```

```
plt.plot(range(100),sgd['train_acc'],label='sgd')
plt.plot(range(100),adagrad['train_acc'],label='adagrad')
plt.plot(range(100),adam['train_acc'],label='adam')
plt.plot(range(100),amsgrad['train_acc'],label='amsgrad')
plt.plot(range(100),adabound['train_acc'],label='adabound',linestyle='dashed')
plt.plot(range(100),amsbound['train_acc'],label='amsbound')
plt.xlabel('epoch')
plt.title('Train Accuracy')
plt.legend()
plt.savefig('1_layer_Perceptron_Train_Acc.png')
```

```
plt.plot(range(100),sgd['val_acc'],label='sgd')
plt.plot(range(100),adagrad['val_acc'],label='adagrad')
plt.plot(range(100),adam['val_acc'],label='adam')
plt.plot(range(100),amsgrad['val_acc'],label='amsgrad')
plt.plot(range(100),adabound['val_acc'],label='adabound',linestyle='dashed')
plt.plot(range(100),amsbound['val_acc'],label='amsbound')
plt.xlabel('epoch')
plt.title('Test Accuracy')
plt.legend()
plt.savefig('1_layer_Perceptron_Test_Acc.png')
```
